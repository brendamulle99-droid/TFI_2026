```
import networkx as nx
import matplotlib.pyplot as plt

# 1. Crear un grafo dirigido (relación direccionada)
G = nx.DiGraph()

# 2. Añadir los 8 nodos requeridos
nodos = [1, 2, 3, 4, 5, 6, 7, 8]
G.add_nodes_from(nodos)

# 3. Definir las conexiones direccionada para asegurar que cada uno se comunique con solo 2 nodos.
aristas = [
    (1, 2), (1, 3),
    (2, 3), (2, 4),
    (3, 4), (3, 5),
    (4, 5), (4, 6),
    (5, 6), (5, 7),
    (6, 7), (6, 8),
    (7, 8), (7, 1),
    (8, 1), (8, 2)
]
G.add_edges_from(aristas)

# 4. Imprimir la información
print("Los vértices de G son: {}".format(list(G.nodes)))
print("Las aristas de G son: {}".format(list(G.edges)))
print("-" * 50)

# Verificamos la comunicación directa
for nodo in G.nodes:
    vecinos_dirigidos = list(G.successors(nodo))
    print("El nodo {} tiene comunicación directa con: {}".format(nodo, vecinos_dirigidos))

# 5. Dibujar la red
plt.figure(figsize=(8, 6)) # Hacemos el grafo un poco más grande
# Usamos un layout circular para que las conexiones se vean claras.
pos = nx.circular_layout(G)

nx.draw(G, pos,
        with_labels=True,
        node_color='lightblue',
        node_size=800,
        font_weight='bold',
        arrows=True, # arrows=True asegura que se vean las flechas de dirección
        arrowsize=20)

plt.title("Red de Comunicación Direccionada (8 nodos)")
plt.show()
```
